## iloc - Index Positions

```python
bond = pd.read_csv("jamesbond.csv")
bond.head()
```

```python
bond.iloc[4]  # Retrieves one particular row
bond.iloc[[15,20]] # Retrieves few particular rows
bond.iloc[4:8] # 4 to 7
bond.iloc[0:6] # 0 to 5
bond.iloc[:6] # 0 to 5
bond.iloc[20:] # 20 till last
bond.iloc[-1]
```

## loc - Index Labels

```python
bond = pd.read_csv("jamesbond.csv",index_col="Film")
bond.head()
```

```python
bond.loc["Goldfinger"] # Single row
bond.loc[["Goldfinger","GoldenEye"]] # Two rows
bond.loc["Diamonds Are Forever":"Moonraker"] # Both the end indices are inclusive
bond.loc["GoldenEye":]
bond.loc[:"GoldenEye"]
```

```python
bond.loc["Casino Royale"] # 2 entries
# bond.loc[:"Casino Royale"] # KEY ERROR
# bond.loc["Casino Royale":] # KEY ERROR
```

## Second Arguments in loc and iloc

```python
# If the index is sorted, then searching becomes easy
bond = pd.read_csv("jamesbond.csv",index_col="Film").sort_index()
bond.head()
```

```python
# Intersection of particular row and column
bond.loc["Diamonds Are Forever", "Director"]
```

```python
bond.loc[["Diamonds Are Forever", "GoldenEye"], "Director"]
```

```python
bond.loc[["Diamonds Are Forever", "GoldenEye"], ["Director", "Year"]]
```

```python
bond.loc[["Diamonds Are Forever", "GoldenEye"], "Director":"Budget"]
```

```python
bond.loc["Diamonds Are Forever": "Octopussy", "Director":"Budget"]
```

```python
# Intersection of 0th row and 2nd column
bond.iloc[0,2]
```

```python
bond.iloc[[0,2],[3,5]]
```

```python
bond.iloc[:7,:3]
```

## Overwrite values in dataframe

```python
bond = pd.read_csv("jamesbond.csv",index_col="Film").sort_index()
bond.head()
```

```python
# We are changing a value in original dataframe - it generates a warning - not recommended
# bond["Actor"].loc["Diamonds Are Forever"] = "Sir Sean Connery"
```

```python
# No warning
bond.loc["Diamonds Are Forever","Actor"] = "Sir Sean Connery"
```

```python
bond
```

## Overwrite multiple values in dataframe

```python
bond = pd.read_csv("jamesbond.csv",index_col="Film").sort_index()
bond.head()
```

```python
# Using "replace function"
bond["Actor"] = bond["Actor"].replace("Sean Connery","Sir Sean Connery")
bond
```

```python
# This will not change the original data frame
bond[bond["Actor"] == "Sean Connery"].loc[:,"Actor"] = "Sir Sean Connery"
bond
```

```python
# This changes the original dataframe
is_sean_connery = bond["Actor"] == "Sean Connery"
bond.loc[is_sean_connery,"Actor"] = "Sir Sean Connery"
bond
```
