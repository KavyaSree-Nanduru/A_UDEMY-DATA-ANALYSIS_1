## Rename columns

```python
import pandas as pd
```

```python
bond = pd.read_csv("jamesbond.csv",index_col="Film").sort_index()
bond.head()
```

```python
# Column labels
bond = bond.rename(columns = {"Year":"Year_of_release", "Box Office":"Revenue"})
bond
```

```python
# Row labels
swaps = {
    "Goldfinger":"Golden_Finger","No Time to Die":"No_Time_to_die"
        }
bond = bond.rename(index = swaps)
bond
```

```python
# To change all column names
bond.columns = ["YEAR","ACTOR NAME","DIRECTOR NAME","REVENUE","ESTIMATE","SALARY"]
bond.head()
```

# Delete rows and columns

```python
bond = pd.read_csv("jamesbond.csv",index_col="Film").sort_index()
bond.head()
```

```python
# Delete columns
bond = bond.drop(columns = ["Box Office","Budget"])
bond.head()
```

```python
# Delete both rows and columns
bond = bond.drop(index = ["A View to a Kill","Casino Royale"], columns = ["Box Office","Budget"])
bond.head()
```

```python
# pop deletes entire column
bond.pop("Actor")
bond.head()
```

```python
del bond["Year"]
bond.head()
```

## nsmallest and nlargest - They work only on numerical columns

```python
bond = pd.read_csv("jamesbond.csv",index_col="Film").sort_index()
bond.head()
```

```python
# Retrieve the 4 films with highest box office gross
bond.sort_values("Box Office", ascending = False).head(4)
```

```python
# Retrieve the 4 films with highest box office gross
bond.nlargest(n=4,columns = "Box Office")
# bond.nsmallest(4,columns = "Box Office")
```

```python
# nlargest can be applied on a series as well
bond["Box Office"].nlargest(4)
# bond["Box Office"].narlest(4)
```

## where method

```python
bond = pd.read_csv("jamesbond.csv",index_col="Film").sort_index()
bond.head()
```

```python
# Original dataframe does not change with "where"
FIND_ACTOR = bond["Actor"] == "Sean Connery"
bond.where(FIND_ACTOR)

# Other ways
# bond.loc[FIND_ACTOR]
# bond[FIND_ACTOR]
```

## apply method

```python
bond = pd.read_csv("jamesbond.csv",index_col="Film").sort_index()
bond.head()
```

```python
bond["Actor"].apply(len)
```

```python
## Movie Ranking System
# CONDITION     DESIGNATION
# 80s movie -- Good_movie
# Pierce Brosnan -- Best_Bond_Movie
# Budget >100 -- "Expensive" and "fun"
# Others -- No_Comment

def rank_movie(row):
    year = row.loc["Year"]
    actor = row.loc["Actor"]
    budget = row.loc["Budget"]
    
    if year >= 1980 and year < 1990:
        return "Good_Movie"
    
    if actor == "Pierce Brosnan":
        return "Best_Bond_Movie"
    
    if budget >100:
        return "Expensive_And_Fun"
    
    return "No_Comment"    

# Function applies for every row
bond.apply(rank_movie, axis = "columns")
#axis = we will traverse across columns and pass rows to the function
```
