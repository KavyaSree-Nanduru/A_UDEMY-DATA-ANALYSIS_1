## Common String Methods

```python
import pandas as pd
```

```python
# Drop the columns which is totally null
c = pd.read_csv("chicago.csv").dropna(how = "all")
```

```python
c.head()
```

```python
c.info()
```

```python
c.nunique()
```

```python
c["Department"] = c["Department"].astype("category")
```

```python
c["Position Title"] = c["Position Title"].str.title()
```

```python
# Positions which have "water" in their title
water_workers = c["Position Title"].str.lower().str.contains("water")
c[water_workers]
```

```python
# Positions that start with "civil"
civil = c["Position Title"].str.lower().str.startswith("civil")
c[civil]
```

## String methods on row labels and column labels

```python
# Drop the columns which is totally null
c = pd.read_csv("chicago.csv",index_col="Name").dropna(how = "all")
c["Department"] = c["Department"].astype("category")
c.head()
```

```python
c.index = c.index.str.strip().str.title()
```

```python
c.head()
```

```python
c.columns = c.columns.str.upper()
```

## Split Method

### The most common first word in the job position title

```python
# Drop the columns which is totally null
c = pd.read_csv("chicago.csv").dropna(how = "all")
c["Department"] = c["Department"].astype("category")
c.head()
```

```python
c["Position Title"].str.split(" ")
```

```python
c["Position Title"].str.split(" ").str.get(0).value_counts()
```

### The most common first name among the employees

```python
# Drop the columns which is totally null
c = pd.read_csv("chicago.csv").dropna(how = "all")
c["Department"] = c["Department"].astype("category")
c.head()
```

```python
# In the name column, after the first word, there is a comma and a space. 
#So we have included a comma and a space in split method
# Now, first name is the second word. So it is at the index position 1
c["Name"].str.title().str.split(", ").str.get(1)
```

```python
# Now we are extracting one value to find the format of the first names
c["Name"].str.title().str.split(", ").str.get(1).iloc[0]
```

```python
# There is white space in the beginning of the word. So we use strip method
c["Name"].str.title().str.split(", ").str.get(1).str.strip()
```

```python
c["Name"].str.title().str.split(", ").str.get(1).str.strip().str.split(" ")
```

```python
c["Name"].str.title().str.split(", ").str.get(1).str.strip().str.split(" ").str.get(0).value_counts()
```

## Expand and n parameters

```python
# Drop the columns which is totally null
c = pd.read_csv("chicago.csv").dropna(how = "all")
c["Department"] = c["Department"].astype("category")
c.head()
```

```python
c[["First_Name","Last_Name"]] = c["Name"].str.split(",", expand=True)
```

```python
c.head()
```

```python
c["Position Title"].str.split(" ", expand=True)
```

```python
c[["Primary_Title","Secondary_Title"]] = c["Position Title"].str.split(" ", expand=True, n=1)
c.head()
```
