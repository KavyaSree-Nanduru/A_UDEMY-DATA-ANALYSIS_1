## Filling missing values

```python
nba = pd.read_csv("nba.csv").dropna(how = "all")
nba
```

```python
# Copy is created and original series is replaced 
nba["Salary"] = nba["Salary"].fillna(0)
nba
```

```python
nba["College"] = nba["College"].fillna("Unknown")
nba
```

```python
nba["Weight"].isnull().any()
```

```python
nba["Weight"] = nba["Weight"].fillna(nba["Weight"].mean())
```

## astype - 1

### If an interger column has null values, pandas automatically converts it into float datatype

```python
nba = pd.read_csv("nba.csv").dropna(how = "all")
```

```python
nba["Salary"] = nba["Salary"].fillna(0)
nba["Weight"] = nba["Weight"].fillna(0)
```

```python
nba.dtypes
```

```python
# We can convert like this, only after filling all the null values. Without that, we get get an error
nba["Salary"] = nba["Salary"].astype(int)
nba["Weight"] = nba["Weight"].astype(int)
```

```python
nba.dtypes
```

## sort_values

```python
nba = pd.read_csv("nba.csv")
nba
```

```python
nba.sort_values(by="Name",ascending = False,na_position="first")
   # na_position -> first or last
```

```python
# First sorted by "Team" and then by "Name" both in descending
nba.sort_values(by=["Team","Name"], ascending = False)
```

```python
# Team is sorted in ascending and then Name is sorted in descending
nba.sort_values(by=["Team","Name"], ascending = [True,False])
```

```python
# Assign it to original dataframe to make permanent changes
nba = nba.sort_values(by=["Team","Name"], ascending = [True,False])
```

## rank method

```python
nba = pd.read_csv("nba.csv").dropna(how = "all")
nba.head()
```

```python
nba["Salary"] = nba["Salary"].fillna(0).astype(int)
```

```python
# Rank 1 is given to highest salary
# Standard return type of rank is float
nba["Salary Rank"] = nba["Salary"].rank(ascending=False).astype(int)
```

```python
nba.sort_values("Salary",ascending=False).head(10)
```
